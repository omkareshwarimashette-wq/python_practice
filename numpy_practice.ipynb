```python
#Q15.2: Use np.where to categorize customers by balance ranges
import numpy as np
balances = bank_data[:, 2]
balance_category = np.where(balances < 50000, "Low",
                    np.where(balances <= 100000, "Medium", "High"))
print(balance_category)
```

```python
#Q15.3: Use np.select to categorize customers by credit score ranges
import numpy as np
credit_scores = bank_data[:, 3].astype(int)
conditions = [
    credit_scores < 600,
    (credit_scores >= 600) & (credit_scores <= 700),
    credit_scores > 700
]
choices = ["Poor", "Average", "Good"]
credit_category = np.select(conditions, choices, default="Unknown")
print(credit_category)
```

```python
#Q15.4: Use np.piecewise to apply different interest rates based on balance
import numpy as np
balances = bank_data[:, 2]
interest = np.piecewise(
    balances,
    [balances < 50000, (balances >= 50000) & (balances <= 100000), balances > 100000],
    [lambda x: x * 0.03, lambda x: x * 0.05, lambda x: x * 0.07]
)
print(interest)
```
